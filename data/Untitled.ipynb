```python
import re
def getBold(para):
    try:
        return para.split('</b>')[0].split('<b>')[1].strip()
    except:
        return ""

def getDesc(para):
    try:
        return para.split('<br/>')[1].strip()
    except:
        return ""

def getTitle(bold):
    try:
        return re.split('\d+', bold.split('(')[0])[1].strip()
    except:
        return ""
    

def getDept(bold):
    try:
        return re.split('\d+', bold.split('(')[0])[0].strip()
    except:
        return ""
    
    
def getCourseNum(bold):
    try:
        return (int)(re.findall('\d+', bold)[0].strip())
    except:
        return -1

def getGenEd(bold):
    try:
        return [s.strip() for s in bold.split('(')[1].split(')')[1].split(',')]
    except:
        return []

def getCredits(bold):
    try:
        return (int)(bold.split('(')[1].split(')')[0][0])
    except:
        return -1
    
```

```python
f = []
for index, row in dept_content.iterrows():
    contents = row.contents.split("</p>")
    i = 0
    for content in contents[0:-1]:
        d = {}
        bold = getBold(content)
        desc = getDesc(content)
        title = getTitle(bold)
        num = getCourseNum(bold)
        gened = getGenEd(bold)
        dept = getDept(bold)
        credits = getCredits(bold)
        if(num != -1):
            d['_id'] = i
            d['dept_abbrev'] = dept
            d['course_number'] = num
            d['course_credits'] = credits
            d['desc'] = desc
            d['title'] = title
            d['QSR'] = str('QSR' in gened).lower()
            d['VLPA'] = str('VLPA' in gened).lower()
            d['INS'] = str('I&amp;S' in gened).lower()
            d['NW'] = str('NW' in gened).lower()
            d['W'] = str('W' in gened).lower()
            d['DIV'] = str('DIV' in gened).lower()
            i += 1
        f.append(d)

df = pd.DataFrame.from_dict(f)
```
